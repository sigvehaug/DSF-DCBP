# Python Syntax

## Variables
A variable is a name that refers to a value, like a box. It is a way to store and manipulate data in a program. In Python, you can create a variable by assigning a value to it using the equals sign (=). For example:
```python
myNum = 4
```

You can use your variables several times. And put something else inside the same variable/box:
```python
myNum = 5
```

We can use variables to store many different types of data:
```python
myInt = 4
myReal = 2.5
myChar = 'a'
myString = "hello"
```

Different data types

We can use the **print()** function to extract and display what is inside each variable.
```python
print(myInt)
print(myReal)
print(myChar, myString) # you can also print multiple things at the same time, you just need to put a comma between them
```

## Comments

Comments start with a hashtag symbol '**#**'. Everything that you put after it in the same line is ignore by Python. It is useful to **you** as programmer, to remember and document what you are doing.

Comments can also come after the code on the same line.

```python
print('hello') # Writes 'hello'
# We can put comments where we want.
print(myString) # This prints the content of myString, that is again 'hello'.
# And all comments are ignored
```

## Single and multiline commands

You can also run multiple commands in a single line.

```python
myNum = 4
print(myNum)
                    # it does not matter if you put empty lines in between

print(myNum)

myNum_doubled = myNum * 2
print(myNum_doubled)
```

And you can also run a single line command in multiple lines, using wither backslash '\' or '()'.

```python
sum = 2 + 3 + 5 \
      - 5*2
print(sum)

sum = (2 + 3 + 5
      - 5*2)
print(sum)
```

Or you can write multiple statements on one line by separating with ';'.
```python
x=5; y=6; print(x-y)
```

## Indentation

Indentation is very important in Python, it defines blocks of code. 

The start of an indentation block is expressed with the ':' symbol. The rest of the code needs to have either some empty spaces or tab spaces afterwards.

Wrong indentation will lead to an error. For example, the following code will not work:

````python
if x > 0:
print("x is positive")
````

```
IndentationError: expected an **indented** block after 'if' statement on line 1
```
In this case for example, we want the print statement to be executed only if the condition is true, that is if x is positive. To do this, we need to indent the print statement, that is to put some spaces in front of it. The convention is to use 4 spaces or a tab space for each indentation level.

Let's write it correctly:

```python
if x > 0:
    print("x is positive")
```

## Round parenthesis

We have already seen one use the the round parenthesis when we used the print() function.

That is called "**calling**". Whenever we put them in front of a function, we are calling and executing that function.

```python
print(myNum) # Calling
L = [4,2,3,1]
sorted(L)  # Calling 
print(L) # Calling
```

Both **print** and **sorted** are 'built-in' functions. That means Python already knows what to do when we call them. You can create your own functions, and we'll see how later.

If you don't know what a function does you can also inspect it by printing the function name **whitout** executing it.
```python
print(sorted) # notice that we didn't put the parenthesis after 'sorted'
```

Parenthesis can be used also for **grouping**. Exactly like in math expressions.
```python
y = 2 * (3+4) # Grouping
```

# Python Semantics - Variables and Objects


## Data types and assignment

As introduced earlier, Python supports many **data types**.
Unlike some languages, you do **not** declare the type of a variable beforehand.

Python uses **dynamic typing**: the interpreter determines the type automatically from the value you assign.

Common basic types:

* **integers (`int`)**  whole numbers
* **floats (`float`)**  real numbers with decimals
* **strings (`str`)**  text
* **lists (`list`)**  ordered collections of objects

---

### Assigning values

We assign a value to a variable using `=`:

```python
my_variable = 5
print(type(my_variable))   # int

my_variable = 5.0
print(type(my_variable))   # float
```

The type depends on the value:

* `5`  integer
* `5.0`  float

The same variable name can later refer to an object of a different type.

## What *really* is a variable in Python?

We often visualize variables as **boxes containing values**.
That model works at the beginning but it is not how Python actually behaves.

Lets refine it.

---

### Where are variables stored?

All values created while your program runs live in the **RAM** (main memory) of your computer.

RAM is *volatile* memory:

* it exists only while the program is running
* it is cleared when the interpreter stops
* restarting the notebook resets everything

That is why after restarting the kernel or turning off your computer, **all variables disappear**. They only existed during execution.

---

### So what is a variable?

A Python variable does **not store a value directly**.

It stores a **reference to an object in memory**.

You can think of it as a label attached to an object rather than a container holding data.

```
x = 5
```

Conceptually:

```
x  [ integer object: 5 ]
```

Important terminology:

* Not a box
* Not exactly a pointer (in the C sense)
* A **reference binding** between a name and an object

This explains many behaviors in Python (shared objects, mutability, function arguments, etc.).

---

### What kind of things do variables refer to?

We previously talked about *data types* such as:

* integers
* floats
* strings

But in Python, almost everything is actually an **object**.

Python is an **object-oriented language**: every value carries both data and behavior.

An object contains:

1. a value (e.g., `5`)
2. a type (`int`)
3. additional functionality attached to it

These functionalities are accessed using the dot operator `.`

---

### Attributes and methods

Objects expose two kinds of features:

- **Attributes**  stored information about the object
- **Methods**  functions that belong to the object

Example:

```python
text = "hello"
text.upper()
```

Here:

* `"hello"` is a string object
* `upper` is a method belonging specifically to strings

General rule:

| Concept   | Meaning                        |
| --------- | ------------------------------ |
| Attribute | property describing the object |
| Method    | action the object can perform  |

Another example:

```python
x = 5.1
print(x.is_integer()) # A method, that returns True if the value inside the object x is an integer and False if it's not.
print(x.real, '+', x.imag) # Two attributes, containing the value for the real and imaginary part respectively.
```


**To summarize:**

A Python variable is:

> a name that refers to an object stored in memory

And an object is:

> a bundle of data and behavior (attributes + methods)

---

Because Python works with objects and resolves types at runtime, it performs more work during execution than compiled languages like C or C++.

To compensate, many performance-critical libraries are implemented in faster languages (C/C++/Rust) and exposed to Python through bindings.
So we write simple Python code while the heavy computation runs in optimized native code behind the scenes.

# Python Operators

## Arithmetic operators

Python supports the standard arithmetic operations:

```python
# addition, subtraction, multiplication
print((4 + 8) * (6.5 - 3))

# true division (always produces a float)
print(11 / 2)

# floor division (integer part only)
print(11 // 2)

# remainder (modulo)
print(11 % 2)

# exponentiation (power)
print(4 ** 0.5)   # square root
```

There also exist **bitwise operators**.
They are less common in everyday Python and can be looked up when needed.

---

## Assignment operators

We can update a variable using shorthand assignment operators:

```python
x = 2
x += 2
print(x)

x *= 2
print(x)

# try others: -=, /=, //=, %=, **=
```

---

## Comparison operators

These operators compare values and return a boolean (`True` or `False`):

```python
print(4 == 6)   # equal
print(4 != 6)   # not equal
print(4 < 6)
print(4 > 6)
print(4 <= 6)
print(4 >= 6)
```

---

## Boolean operators

Used to combine logical conditions:

* `and`
* `or`
* `not`
* XOR (no dedicated keyword in Python)

```python
x = 4

print((x < 6) and (x > 2))
print((x > 10) or (x % 2 == 0))
print(not (x < 6))

# XOR: exactly one condition must be true
print((x > 1) != (x < 10))
```

---

## Identity and membership operators

Used to check object identity and container membership:

* `is`, `is not`  same object in memory
* `in`, `not in`  element inside a collection

```python
print(1 is 2)
print(1 is 1)

print(1 in [1, 2, 3])
print(2 not in [1, 2, 3])
```

# Python Built-In Data Structures

Python provides **container types** that store multiple objects inside a single variable.

| Type    | Example            | Description                             |
| ------- | ------------------ | --------------------------------------- |
| `list`  | `[1, 2, 3]`        | Ordered, mutable collection             |
| `tuple` | `(1, 2, 3)`        | Ordered, immutable collection           |
| `dict`  | `{'a': 1, 'b': 2}` | Key  value mapping                     |
| `set`   | `{1, 2, 3}`        | Unordered collection of unique elements |

Notice the brackets matter:

* `[]`  list
* `()`  tuple
* `{}`  dictionary or set (depending on content)

---

## Lists

Lists are ordered and can be modified.

```python
L = [2, 3, 5, 7]

print(len(L))        # number of elements

L.append(11)         # add element at the end
print(L)

print(L + [13, 17])  # concatenation creates a new list

L = [2, 5, 1, 6, 3, 4]
L.sort()             # sorts in-place
print(L)

# lists can contain different types
L = [1, 'two', 3.14, [0, 3, 5]]
print(L)
```

---

## Indexing and slicing

Lists can access individual elements or ranges.

```python
L = [2, 4, 5, 'madonna']

print(L[0])      # first element
print(L[-1])     # last element

print(L[2:4])    # slice (start:stop)

print(L[3:])     # from index 3 to end
print(L[:])      # copy of list
print(L[0:-1:2]) # step size = 2
print(L[::-1])   # reversed list
```

---

## Tuples

Tuples are like lists but **cannot be changed after creation**.

```python
t = (2, 3, 5, 7, 11)

print(t[0])

# try modifying it
t[0] = 3
```

*(You should get an error  this is expected.)*

## Dictionaries

Dictionaries map **keys to values**.

```python
numbers = {'one': 1, 'two': 2, 'three': 3}

print(numbers['two'])   # access value by key

numbers['car'] = 'Auto' # add new key
print(numbers)

print(numbers.get('car'))
```

Python documentation lists many useful dictionary methods:
[https://docs.python.org/3/library/stdtypes.html](https://docs.python.org/3/library/stdtypes.html)

---

## Sets

Sets store unique elements and support set operations.

```python
primes = {2, 3, 5, 7}
odds = {1, 3, 5, 7, 9}

print(primes.union(odds))
print(primes.intersection(odds))
print(primes - odds)
```

---

## Empty structures

Sometimes you need an empty container first:

```python
L = []
t = ()
d = {}
s = set()

print(L, t, d, s)
```

# Python ``del`` statement

The syntax of the ``del`` statement is:

```python
del object_name
```
where ``object_name`` can be variable(s), user-defined objects (also classes, attributes of classes), lists, items within lists, dictionaries etc.

**``del`` removes variable names from the namespace**

As a result of using del one can free up for example some variable names from the namespace. This command doesn't necessarily affect the object associated to the variable name, it frees up just the pointer (so that it can be used for other purposes maybe). However in some memory intensive applications it might be also used to prepare the objects to be erased by the garbage collector of Python.
