One normally distinguishes between scripting languages, like R or python, and non-scripting languages like C/C++ and Fortran. Scripts can be executed immediately statement by statement during runtime and do not require compilation, i.e. the explicit translation step into machine code.

```python
# This little method, also called function, in python which
# implements Euclid's algorithm for greatest common divisors

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(gcd(48, 18))
print(gcd(100, 35))
```

### Clean and split the data

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# dropping time is only important if two movements have been recorded one after the other
training = movement.drop(columns=['Time (s)'])
train, test = train_test_split(training, train_size=0.8)
print(train.shape, test.shape)
train
```

## Train and Evaluate Models

Here we try two different models and check for overfitting.

```python
X   = train.drop(columns='type')
y   = train['type']

model = LogisticRegression()
#model = RandomForestClassifier(max_depth=12, random_state=0)
model.fit(X, y)

print(confusion_matrix(train['type'],model.predict(X=train.drop(columns='type'))))
print(confusion_matrix(test['type'], model.predict(X=test.drop(columns='type'))))
print(classification_report(test['type'], model.predict(X=test.drop(columns='type'))))
print(classification_report(train['type'], model.predict(X=train.drop(columns='type'))))
```

```python
print(confusion_matrix(test['type'], model.predict(X=test.drop(columns='type'))))
print(classification_report(test['type'], model.predict(X=test.drop(columns='type'))))
```
